```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_gdp_prices =  pd.read_csv('ab40c054-5031-4376-b52e-9813e776f65e.csv' )
df_gdp_prices
```

##Remove the rows: 
'(% Growth over the previous year)' and 'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17.

```python
df_gdp_prices = df_gdp_prices.drop([5,10])
df_gdp_prices = df_gdp_prices.drop('West Bengal1', axis=1)
```

## Clean the data
Here instead of adding zero , mean value makes more sense

```python
df_inter = df_gdp_prices.iloc[[5,6,7,8]]
df_inter.fillna(df_inter.mean())
```

## Calculate the average growth of states over the duration 2013-14, 2014-15 and 2015-16

```python
df_gdp_gwth_prices = df_inter.iloc[[1,2,3]].mean(axis=0)
df_gdp_gwth_prices = df_gdp_gwth_prices.reset_index()
df_gdp_gwth_prices.columns = ['State', 'Average GDP growth prices']
df_gdp_gwth_prices.drop([28, 29, 30, 31, 32])
df_gdp_gwth_prices.sort_values(['Average GDP growth prices'], ascending=[1], inplace=True)
df_gdp_gwth_prices
```

```python
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Average GDP growth prices', y='State', data=df_gdp_gwth_prices)
ax.set_xlabel('Average GDP growth prices')
```

# Plot the total GDP of the states for the year 2015-16

```python
df_inter_totgdp = df_gdp_prices.iloc[[0,1,2,3,4]]
df_inter_totgdp = df_inter_totgdp.fillna(df_inter_totgdp.mean())
```

```python
df_total_gdp = df_inter_totgdp.iloc[[4]].sum(axis=0)
df_total_gdp = df_total_gdp.reset_index()
df_total_gdp = df_total_gdp.drop([0,1,30,31,32,33, 34, 35])
df_total_gdp.columns = ['State', 'GSDP (in Crore)']
df_total_gdp['GSDP (in Crore)'].astype(int)
df_total_gdp.sort_values(['GSDP (in Crore)'], ascending=[1], inplace=True)
df_total_gdp
```

```python
plt.figure(figsize=(20,10))
ax = sns.barplot(x='GSDP (in Crore)', y='State', data=df_total_gdp)
ax.set_xlabel('GSDP (in Crore)')
```

### Plotting the correlation of primary/upper primary/Secondary education dropout using scatter plot

```python
# Plot

sns.set_style("whitegrid")
sns.jointplot(droprate_gdp_df['Primary'], droprate_gdp_df['Percapita GDP in Rs.'])
plt.show()


sns.jointplot(droprate_gdp_df['Upper Primary'], droprate_gdp_df['Percapita GDP in Rs.'])
plt.show()


sns.jointplot(droprate_gdp_df['Secondary'], droprate_gdp_df['Percapita GDP in Rs.'])
plt.show()
```

### Plotting the correlation of primary/upper primary/Secondary education dropout using scatter plot with Matplotlib. As I ws getting version error in seaborn

```python
plt.figure(figsize=(20,10))

colors = (0,0,0)
area = np.pi*3

# Plot
plt.subplot(2, 2, 1)
plt.scatter(droprate_gdp_df['Primary'], droprate_gdp_df['Percapita GDP in Rs.'], s=area, c=colors, alpha=0.5)
plt.xlabel('Primary')
plt.ylabel('Percapita GDP')

plt.subplot(2, 2, 2)
plt.scatter(droprate_gdp_df['Upper Primary'], droprate_gdp_df['Percapita GDP in Rs.'], s=area, c=colors, alpha=0.5)
plt.xlabel('Upper Primary')
plt.ylabel('Percapita GDP')

plt.subplot(2, 2, 3)
plt.scatter(droprate_gdp_df['Secondary'], droprate_gdp_df['Percapita GDP in Rs.'], s=area, c=colors, alpha=0.5)
plt.xlabel('Secondary')
plt.ylabel('Percapita GDP')
```

### Plotting the correlation of primary/upper primary/Secondary education dropout using HeatMap
